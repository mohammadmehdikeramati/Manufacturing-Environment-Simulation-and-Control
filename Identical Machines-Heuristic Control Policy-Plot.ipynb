```python
# Create a plot for each machine

import matplotlib.pyplot as plt 
import pandas as pd

fig_test, ax_test = plt.subplots(figsize=(12, 6), dpi=1000)

Imported_Data = pd.read_excel("Data.xlsx", sheet_name="Data")

# Define color mapping for states
color_map = {"standby": "gray","busy": "blue","idle": "green",  "failed": "red","startup": "yellow"}

# Plot state transitions for each machine
for machine_id in Imported_Data["machine_id"].unique():
    machine_data = Imported_Data[Imported_Data["machine_id"] == machine_id]
    for i in range(len(machine_data) - 1):
        start_time = machine_data.iloc[i]["time"]
       
        end_time = machine_data.iloc[i + 1]["time"]
        state = machine_data.iloc[i]["state"]
        ax_test.plot([start_time, end_time], [machine_id, machine_id], color=color_map[state], linewidth=4)


ax_test.set_yticks(Imported_Data["machine_id"].unique())
ax_test.set_yticklabels([f"Machine {i}" for i in Imported_Data["machine_id"].unique()])
ax_test.set_xlabel("Time (s)")
ax_test.set_ylabel("Machine ID")
ax_test.set_title("Machine State Transitions Over Time")
ax_test.grid(True)

# Add legend
custom_lines = [plt.Line2D([0], [0], color=color_map[state], lw=4) for state in color_map]
ax_test.legend(custom_lines, list(color_map.keys()))

# Display the plot
plt.show()
```
